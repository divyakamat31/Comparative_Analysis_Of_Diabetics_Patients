```python
#imporing the required library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
#comparative analysis of diabetic patients
```

```python
#uploding the CSV dataset and reading
dataset=pd.read_csv("/content/KNN_Dataset.csv")
```

```python
#length of dataset
len(dataset)
```

```python
#getting first 5 rows using head
dataset.head()
```

```python
#replace zeros
zero_not_accepted=['Glucose','BloodPressure','SkinThickness','BMI','Insulin']
```

```python
#replace by mean value
for column in zero_not_accepted:
 dataset[column]=dataset[column].replace(0,np.NaN)
 mean=int(dataset[column].mean(skipna=True))
 dataset[column]=dataset[column].replace(np.NaN,mean)
```

```python
#row, column segregation
X=dataset.iloc[:,0:8]
y=dataset.iloc[:,8]
X_train,X_test,y_train,y_test=train_test_split(X,y, random_state=0,test_size=0.2)
```

```python
#converting to standard form n storing
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
#we can change it to any type of distance measures
classifier=KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')
```

```python
#applying classifier to the data
classifier.fit(X_train,y_train)
KNeighborsClassifier(algorithm='auto',leaf_size=30, metric='euclidean',metric_params=None,n_jobs=1,n_neighbors=11)
```

```python
#predicting using classifiers
y_pred=classifier.predict(X_test)
y_pred
```

```python
#getting confusion matrix for f1score
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(f1_score(y_test,y_pred))
```

```python
#getting confusion matrix for accuracy
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```
